What are the Top selling products in the super store ??

```python
product_group=sup.groupby('Product Name',as_index=False)['Sales'].sum().sort_values('Sales',ascending=False)[:10]
```

```python
product_group.head()
```

```python
px.bar(data_frame=product_group,x='Product Name',y='Sales')
```

top profit products ?

```python
top10Profit=sup.groupby('Product Name',as_index=False)['Profit'].sum().sort_values('Profit',ascending=False)[:10]
```

```python
px.bar(data_frame=top10Profit,x='Product Name',y='Profit')
```

```python
product_group.index==top10Profit.index
```

```python
fig, (ax1,ax2) =plt.subplots(1,2 ,figsize=(15,5))
product_group.plot(kind='bar',y='Sales', ax=ax1)
ax1.set_title('top 10 selling product')

top10Profit.plot(kind='bar',y='Profit',ax=ax2)
ax2.set_title('top 10 products profit ')

plt.show()
```

```python
multivar=sup[['order date','city','sales']]
multivar=multivar.set_index('order date')
multivar.head()
```

```python
multivar=multivar.resample('1M').agg({'city':'nunique','sales':'sum'})
multivar.head()
```

```python
fig=px.bar(data_frame=multivar,x=multivar.index,y=round(multivar.sales,0),color=multivar.city,text_auto=True)
fig.update_traces(textposition='outside')
fig.update_layout()
fig.show()
```
